```python
def rotate_box(box, w, h, rotation):
    ret_box = [0, 0, 0, 0]
    #print("box:", box, w, h, rotation)
    if rotation == 180:
        ret_box[0] = w-box[2]
        ret_box[2] = w-box[0]
        ret_box[1] = h-box[3]
        ret_box[3] = h-box[1]
    elif rotation==90:
        ret_box[0] = w-box[3]
        ret_box[2] = w-box[1]
        ret_box[1] = box[0]
        ret_box[3] = box[2]
    elif rotation==270:
        #print("################ 270")
        ret_box[0] = box[1]
        ret_box[2] = box[3]
        ret_box[1] = h-box[2]
        ret_box[3] = h-box[0]

    #print("ret_box:", ret_box)
    col_s = min(ret_box[0], ret_box[2])
    col_e = max(ret_box[0], ret_box[2])
    row_s = min(ret_box[1], ret_box[3])
    row_e = max(ret_box[1], ret_box[3])
    #print(row_s, col_s, row_e, col_e)
    return row_s, col_s, row_e, col_e
```

```python
class CioccaDataset(Dataset):
    #load the dataset definitions
    def load_dataset(self, dataset_dir, is_train=True):
        class_id_map_file = "class_id_codes_map.json"
        f = open(class_id_map_file)
        class_codes = json.load(f)
        for class_id in class_codes:
            #print(class_id, class_codes[class_id])
            self.add_class("dataset" , class_id, class_codes[class_id])

        #images_dir = dataset_dir + '/UNIMIB2016-images/original' #TODO: use raw images instead of annotated images
        #images_dir = dataset_dir + '/original' #TODO: use raw images instead of annotated images
        #images_dir = dataset_dir + '/UNIMIB2016-images/original_rotated' #TODO: use raw images instead of annotated images
        #images_dir = dataset_dir + '/annotated_imgs-20220331T112805Z-001/annotated_imgs' #TODO: use raw images instead of annotated images
        images_dir = '/content/original' #TODO: use raw images instead of annotated images

        #json_annot_file = dataset_dir + "/annotations_clean.json"
        json_annot_file = "annotations_clean.json"

        f = open(json_annot_file)
        data = json.load(f)

        self.boxes = {} #image_name : ([boxes], width, height, [class_code]) ... each box as xmin, ymin, xmax, ymax... width and height for whole image
        #self.masks = {}

        #image count with prefix 201511*.jpg = 284 - to be used for testing
        #Count of rest of the images (that happen to be with prefix 201512*.jpg) = 709 - to be used for training
        #TODO: image_id should be generated as unique id based on image_name - instead of depending on order
        image_id = -1
        for image in data:
            image_id+=1
            image_name = image['image_name'][len("original/"):]
            if is_train and '201511' in image_name:
                continue
            if not is_train and '201512' in image_name:
                continue
            #print(image_name, "#", image_id)
            img_path = images_dir + "/" + image_name
            img = PIL.Image.open(img_path)
            width, height = img.size
            self.add_image('dataset', image_id=image_id, path=img_path, annotation=json_annot_file+"#"+image_name)

            #self.boxes[image_name] = image['bounding_box']
            #print("boxes:", image['bounding_box'])
            boxes = image['bounding_box']
            all_box_extents = []
            for box in boxes:
                xmin = min(box[::2])
                ymin = min(box[1::2])
                xmax = max(box[::2])
                ymax = max(box[1::2])
                box_extents = [xmin, ymin, xmax, ymax]
                all_box_extents.append(box_extents)
                #print(xmin, ymin, xmax, ymax)
            #print(image_name, " ", image['bounding_box'])
            #self.masks[image_id] = image['contours']
            #print(image_id, " ", self.masks[image_id])
            
            class_codes = image['class_codes']
            
            self.boxes[img_path] = [all_box_extents, width, height, class_codes] #using img_path instead of image_name


	# extract bounding boxes from an annotation file
    def extract_boxes(self, filename):
        #print("extract_boxes:", self.boxes[filename])
        return self.boxes[filename]

    # load the masks for an image
    def load_mask(self, image_id):
        #print("load mask:", image_id)
        #print("load_mask:", self.masks.keys())
        info = self.image_info[image_id]
        #print("info:", info)
        path = info['path']
        boxes, w, h, class_codes = self.extract_boxes(path)

        #print("!@#!@#:", w, h)
        img_temp = self.load_image(image_id)
        h,w,d = img_temp.shape
        #print("!@#!@#:", w, h)
        rotation = 0
        if w < h:
            rotation = 90
        else:
            rotation = 180

        #print("boxes, w, h = ", boxes, w, h)
        masks = zeros([h, w, len(boxes)], dtype='uint8')
        class_ids = list()
        for i in range(len(boxes)):
            box = boxes[i]
            #row_s, row_e = box[1], box[3]
            #col_s, col_e = box[0], box[2]
            row_s, col_s, row_e, col_e = rotate_box(box, w, h, rotation)
            #print("masking", row_s, row_e, col_s, col_e, i)
            masks[int(row_s):int(row_e), int(col_s):int(col_e), i] = 1
        for class_code in class_codes:
            class_ids.append(self.class_names.index(class_code)) #TODO: actually class names to be used
        return masks, asarray(class_ids, dtype='int32')
    
    def image_reference(self, image_id):
        #print("image_references:", image_id)
        info = self.image_info[image_id]
        return info
```
